```python
svm = SVC(class_weight='balanced',gamma='scale')

svm.fit(X_train_airport,y_train_airport)
y_pred_airport = svm.predict(X_test_airport)
display(pd.crosstab(y_test_airport, y_pred_airport, rownames=['reality_airport'], colnames=['predictions_airport']))
print(classification_report(y_test_airport, y_pred_airport))
```

```python
print(np.round(pd.crosstab(y_test_airport, y_pred_airport,
                           rownames=['real/pred'],
                           normalize=True),2).to_markdown())
print(np.round(pd.DataFrame(
    sklearn.metrics.classification_report(y_test_airport, y_pred_airport,
                                          output_dict=True)),2).to_markdown())
```

```python
lr = LogisticRegression(class_weight = 'balanced')
param = {'solver':['lbfgs', 'liblinear', 'sag'], 'C':[0.001, 0.1, 0.2, 0.5, 1, 10, 20, 50]}
grid_lr = GridSearchCV(estimator = lr, param_grid = param)



grid_lr.fit(X_train_airport, y_train_airport)
print(grid_lr.best_params_)
y_pred_airport = grid_lr.predict(X_test_airport)
display(pd.crosstab(y_test_airport, y_pred_airport, rownames = ['Reality_airport'], colnames = ['Prediction_airport']))
print(classification_report(y_test_airport, y_pred_airport))
```
